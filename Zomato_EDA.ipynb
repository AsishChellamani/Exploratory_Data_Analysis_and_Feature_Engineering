```python
final_df = pd.merge(df,df_country, on='Country Code', how='left')
final_df.head(2)
```

## Exploratory Data Analysis

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
## Pie Chart - Top 3 Countries that uses Zomato
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
```

##### Observation: Zomato maximum records or transactions are from India afetr that USA and UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

## Oberservation
1. Ratings varying from 4.9 to 4.5 indicates that the rating was excellent
2. Ratings varying from 4.4 to 4.0 indicates that the rating was very good
3. Ratings varying from 3.9 to 3.5 indicates that the rating was good
4. Ratings varying from 3.4 to 2.5 indicates that the rating was avergae
5. Ratings varying from 2.4 to 1.8 indicates that the rating was poor
6. Ratings marked as 0.0 indicates towards not rated

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue='Rating color', palette=['grey','red','orange','yellow','green','green'])
```

## Observations
1. Not rated count is very high
2. Maximum no of rating are between 2.5 to 3.4


```python
## Count plot
sns.countplot(x='Rating color',data=ratings,palette=['grey','red','orange','yellow','green','green'])
```
